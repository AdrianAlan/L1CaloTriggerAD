```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('../plots/physics.mplstyle')
```

```python
in_file = '/eos/user/a/adpol/L1AD/Signal/120X/GluGluHToTauTau_M-125_TuneCP5_14TeV.h5'
hdf5_dataset = h5py.File(in_file, 'r')
```

```python
calo_regions = hdf5_dataset['CaloRegions']
```

```python
bins = np.argwhere(calo_regions)
phi, eta = bins[:, 1], bins[:, 2]
et = calo_regions[:][calo_regions[:] != 0]
```

```python
fig = plt.figure(figsize=(14, 3))

ax = plt.subplot(131)
ax.hist(eta + 4,
        density=True,
        facecolor=None,
        bins=np.arange(4, 19),
        histtype='step',
        edgecolor='black')
ax.set_ylabel('a.u.')
ax.set_xlabel(r'i$\eta$')

ax = plt.subplot(132)
ax.hist(phi,
        density=True,
        facecolor=None,
        bins=np.arange(19),
        histtype='step',
        edgecolor='black')
ax.set_ylabel('a.u.')
ax.set_xlabel(r'i$\phi$')

ax = plt.subplot(133)
ax.hist(et,
        bins=50,
        density=True,
        facecolor=None,
        histtype='step',
        edgecolor='black')
ax.set_ylabel('a.u.')
ax.set_xlabel(r'iE$_T$')
ax.set_yscale('log')

fig.suptitle('Dataset Profiling')
plt.show();
```
